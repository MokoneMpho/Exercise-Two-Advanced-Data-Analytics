```sql
create table if not exists dim_table_customers
( customer_id BIGINT primary key,
first_name varchar (50) not null,
 last_name string not null,
 email  string, 
 tier string,
loyalty_points int

);
```

Question 2 create the Dimension (DIM) Products Table (include the Primary Key)

```sql
create table if not exists dim_product_table
(product_id  BIGINT primary key,
product_code string not null,
product_name BIGINT,
price decimal (8,2),
category string,
is_active boolean
);
```

Question 3 Create the fact_order table, partioned by order date (include primary key on order_id, End wit using DELTA PARTITIONED BY order_date)

```sql
create table if not exists fact_order_table
( order_id BIGINT primary key,
 customer_id BIGINT,
 product_id BIGINT,
 store_id BIGINT,
  order_date BIGINT,
quantity INT,
unit_price decimal (8,2),
total_amount decimal (10,2)
)
using delta
partitioned by (order_date);
```
